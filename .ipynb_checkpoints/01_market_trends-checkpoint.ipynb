## üìä Cameron County Real Estate Market Trends (2020‚Äì2024)
This notebook presents key real estate investment data for Cameron County using market trends from 2020 to 2024. - Gilbert J. Realtor


```python
### üîç Cell 1: Import libraries & verify data folder

import os
import pandas as pd
```

### üîç Cell 2: Verify location of datasets

```python
# Detect dataset folder
pdf_folder = "../datasets" if os.getcwd().endswith("notebooks") else "datasets"
files = os.listdir(pdf_folder)

# List PDFs
print("PDFs found:")
for f in files:
    if f.endswith(".pdf"):
        print(f"- {f}")
```

### üìà Cell 3: Manually entered market stats

```python
# Key statistics per year
data = {
    "Year": [2020, 2021, 2022, 2023, 2024],
    "Median_Price": [173000, 215000, 249900, 260000, 260000],
    "Median_Price_SqFt": [105.25, 124.62, 136.75, 140.00, 140.00],
    "Months_Inventory": [4.9, 2.6, 2.4, 2.6, 3.0],
    "Days_on_Market": [129, 98, 80, 94, 114]
}

# Create DataFrame
os.makedirs("../processed", exist_ok=True)
df = pd.DataFrame(data)
df.to_csv("../processed/market_stats.csv", index=False)
df.head()
```

### üìä Cell 4: Median Price Trend Line


```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(df["Year"], df["Median_Price"], marker='o', linewidth=2, color='blue')
plt.title("Median Price Growth in Cameron County (2020‚Äì2024)", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Median Price ($)")
plt.grid(True)
plt.tight_layout()
plt.savefig("../visuals/median_price_trend.png")
plt.show()
```

### üìä Cell 5: Dual-Axis Chart for Median Price & Price per SqFt


```python
fig, ax1 = plt.subplots(figsize=(12, 6))

# Median Price - Left Axis
ax1.plot(df["Year"], df["Median_Price"], color='blue', marker='o', linewidth=2, label="Median Price ($)")
ax1.set_xlabel("Year")
ax1.set_ylabel("Median Price ($)", color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Annotate Median Prices
for x, y in zip(df["Year"], df["Median_Price"]):
    ax1.annotate(f"${y:,}", (x, y + 3000), color='blue', fontsize=9, ha='center')

# Price per SqFt - Right Axis
ax2 = ax1.twinx()
ax2.plot(df["Year"], df["Median_Price_SqFt"], color='orange', marker='s', linestyle='--', linewidth=2, label="Price per SqFt ($)")
ax2.set_ylabel("Price per SqFt ($)", color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Annotate SqFt prices
for x, y in zip(df["Year"], df["Median_Price_SqFt"]):
    ax2.annotate(f"${y:.2f}", (x, y + 2), color='orange', fontsize=9, ha='center')

# Appreciation Arrow
growth = df["Median_Price"].iloc[-1] - df["Median_Price"].iloc[0]
pct = round((growth / df["Median_Price"].iloc[0]) * 100, 1)
ax1.annotate(f"‚Üë ${growth:,} / {pct}% gain",
             xy=(df["Year"].iloc[-1], df["Median_Price"].iloc[-1]),
             xytext=(df["Year"].iloc[-2], df["Median_Price"].iloc[-1] - 15000),
             arrowprops=dict(facecolor='green', shrink=0.05),
             fontsize=10, color='green',
             bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="green", lw=1))

plt.title("Cameron County Real Estate Investment Growth (2020‚Äì2024)", fontsize=14)
fig.tight_layout()
os.makedirs("../visuals", exist_ok=True)
plt.savefig("../visuals/investor_dual_axis_labeled.png")
plt.show()
```

### üìä Cell 6: DOM vs Inventory Trend

```python
fig, ax1 = plt.subplots(figsize=(12, 6))

# Days on Market - Left
ax1.plot(df["Year"], df["Days_on_Market"], marker='o', color='red', label="Days on Market", linewidth=2)
ax1.set_ylabel("Average DOM", color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Months Inventory - Right
ax2 = ax1.twinx()
ax2.plot(df["Year"], df["Months_Inventory"], marker='s', linestyle='--', color='green', label="Months of Inventory", linewidth=2)
ax2.set_ylabel("Months of Inventory", color='green')
ax2.tick_params(axis='y', labelcolor='green')

plt.title("Market Speed & Inventory (2020‚Äì2024)", fontsize=14)
fig.tight_layout()
plt.savefig("../visuals/dom_inventory_dual_axis.png")
plt.show()
```
